```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.utils import to_categorical
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.image as mpimg
import cv2
import os
import random
import pandas as pd

%matplotlib inline
```

```python
def display_random_example(train_path):
    plt.figure(figsize=(20, 20))
    path = fr'{train_path}'
    for i in range(5):
        file = random.choice(os.listdir(path))
        image_path = os.path.join(path, file)
        img = mpimg.imread(image_path)
        ax=plt.subplot(1, 5, i+1)
        ax.title.set_text(file)
        plt.imshow(img)

        
display_random_example((os.getcwd() + '/Test'))
```

```python
def create_dataset(img_folder, img_width, img_height):
    data = []
    labels = []
    
    for dir1 in os.listdir(img_folder):
        for file in os.listdir(os.path.join(img_folder, dir1)):
            image_path = os.path.join(img_folder, dir1, file)
            image = cv2.imread(image_path, cv2.COLOR_BGR2RGB) # wczytanie obrazkow
            image = cv2.resize(image, (img_height, img_width), interpolation=cv2.INTER_AREA)
            image = np.array(image)  
            image = image.astype('float32')
            image /= 255  #znormalizowanie kolorow od 0 do 1
            data.append(image)
            labels.append(dir1)
    return data, labels

img_data, img_labels = create_dataset((os.getcwd() + '/train'), 30, 30)
img_labels_int = list(map(int, img_labels))
#img_labels_int.sort()
y, x, _ = plt.hist(img_labels_int, [x for x in range(0, 43)])
print(y)
print(y.max())
print(y.min())
mean = np.mean(y)
print(mean)
std = np.std(y)
print(std)
img_data = np.array(img_data)
img_labels = np.array(img_labels)


# print(img_data.shape, img_labels.shape)
# print(np.unique(img_labels))
```

```python
X_train, X_test, y_train, y_test = train_test_split(img_data, img_labels, test_size=0.2)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
y_train = to_categorical(y_train, 43)
y_test = to_categorical(y_test, 43)
print(y_train)
```

```python
def display_info_sample(num):
    print(y_train[num])
    label = y_train[num].argmax(axis=0)
    image = X_train[num]
    plt.title('Sample: %d Label: %d' % (num, label))
    plt.imshow(image, cmap=plt.get_cmap('gray_r'))
    plt.show()
display_info_sample(120)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(30, 30, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(43, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', #categorical_crossentropy poniewaz robimy wielokronta kategoryzacje
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    batch_size=32,
                    epochs=20,
                    verbose=2,
                    validation_data=(X_test, y_test))
```

```python
y_test = pd.read_csv('Test.csv')
img_path = y_test['Path'].values
img_class = y_test['ClassId'].values
images = []
img_width = 30
img_height = 30
for image in img_path:
    img = cv2.imread(image, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (img_height, img_width), interpolation=cv2.INTER_AREA)
    img = np.array(img)
    images.append(img)
    
X_test = np.array(images)
prediction = model.predict_classes(X_test)
print(prediction)
accuracy_score(img_class, prediction)
```

```python
plt.plot(history.history['val_loss'], label='val_loss')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
```
